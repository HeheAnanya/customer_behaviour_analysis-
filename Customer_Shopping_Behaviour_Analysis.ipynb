```python
df["Review Rating"] = df.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median()) ) #One of the ways
```

```python
group_medians = df.groupby("Category")["Review Rating"].median()
df["Review Rating"] = df["Review Rating"].fillna(df["Category"].map(group_medians))
```

df.isnull().sum()

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_")
df
#Here we HAVE CONVERTED OUR COLUMN NAMES INTO SNAKE CASING
```

```python
df = df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
df
```

```python
#CREATING A COULMN FOR AGE GROUPS like teenager, adult etc 
categories = ["Young_Adult","Adult","Middle_Age","Senior"]
df["age_groups"] = pd.qcut(df.age, q=4,labels=categories)
```

```python
df[["age","age_groups"]].head(20)
```

```python
# CREATING COLUMN  PURCHASE_FREQUNCY_DAYS
df.frequency_of_purchases.value_counts()
frequency_mapping = {
    "Every 3 Months":90,
    "Annually":365,
    "Quarterly":90,
    "Monthly":30,
     "Bi-Weekly":14,
    "Fortnightly":14,
    "Weekly":7      
}
df["purchase_frequency_days"] = df.frequency_of_purchases.map(frequency_mapping)
df.purchase_frequency_days.head(5)
```

```python
(df.discount_applied==df.promo_code_used).all()
#If TRUE-> we will drop either else not
```

```python
df.drop(columns=["promo_code_used"], inplace=True)
```
